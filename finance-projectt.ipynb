```python
from pandas_datareader import data as web
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sns


# plotly packages
import plotly
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)
cf.go_offline()


%matplotlib inline
```

```python
df = pd.read_pickle('/kaggle/input/finance-data-of-banks/all_banks')
```

```python
# 1. Bank of America data 
BAC = df['BAC']
BAC
```

```python
# 2. CitiGroup data 
C = df['C']
C
```

```python
# 3. Goldman Sachs data 
GS = df['GS']
GS
```

```python
# 4. JPMorgan Chase data 
JPM = df['JPM']
JPM
```

```python
# 5. Morgan Stanley data 
MS = df['MS']
MS
```

```python
# 6. Wells Fargo data 
WFC = df['WFC']
WFC
```

```python
# tickers list of all banks 
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC' ]
```

```python
# all banks data 
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys=tickers)
```

```python
bank_stocks.info()
```

```python
bank_stocks.head()
```

```python
# setting columns names
bank_stocks.columns.names = ['Bank ticker', 'Stock info']
```

```python
bank_stocks.head()
```

```python
# one way of getting BAC max stock price
bank_stocks['BAC']['Close'].max()
```

```python
# one way of getting all stocks max price
for ticker in tickers:
    print(ticker, bank_stocks[ticker]['Close'].max())
    
```

```python
# betterb way of getting the same data, using sc() method, (cross-section)
bank_stocks.xs('Close', axis=1,level='Stock info').max()
```

```python
# creating new data frame
returns = pd.DataFrame()
```

```python
# daily % return for BAC bank
BACreturns = bank_stocks['BAC']['Close'].pct_change()
BACreturns.head()
```

```python
# Daily returns for every bank
for ticker in tickers:
    returns[ticker+' Return'] = bank_stocks[ticker]['Close'].pct_change()
    
```

```python
returns
```

```python
sns.pairplot(data=returns)
```

```python
# worst single day return for each bank
min = returns.min()

# the date of the worst single day return for each bank
idmin = returns.idxmin()
```

```python
pd.concat([min, idmin], axis=1)
```

```python
# best single day return for each bank
max = returns.max()

# the date of the best single day return for each bank
idmax = returns.idxmax()
```

```python
pd.concat([max, idmax], axis=1)
```

```python
# standard deviation of stocks
std = returns.std()
std
```

```python
# standard deviation of stocks in 2015
returns.loc['2015-01-01':'2015-12-31'].std()
```

```python
# Distribution of Morgan Stanley stocks in 2015, (x = standard deviation )
sns.displot(returns.loc['2015-01-01':'2015-12-31']['MS Return'], bins=50, kde=True, color='green' )

```

```python
# Distribution of CityGroup stocks in 2008, (x = standard deviation )

sns.displot(returns.loc['2008-01-01':'2008-12-30']['C Return'], bins=50, kde=True, color='red' )
```

```python
# Creating a plot showing Close price for the entire index time (for loop)
for ticker in tickers:
    
    bank_stocks[ticker]['Close'].plot(label=ticker, figsize=(15,6))
```

```python
# Creating a plot showing Close price for the entire index time (xs() method)
bank_stocks.xs(key='Close', axis=1,level='Stock info').plot(label=ticker, figsize=(25,10))
```

```python
# absolutly the best way of showing the same data, using cufflings
bank_stocks.xs(key='Close', axis=1,level='Stock info').iplot()
```

```python
# 2 methodes of showing BAC Close Price and 30 day moving average 

# 1.
# bank_stocks.loc['2008-01-01':'2008-12-30']['BAC']['Close'].plot( label='BAC Close Price', figsize=(25,10))
# bank_stocks.loc['2008-01-01':'2008-12-30']['BAC']['Close'].rolling(30).mean().plot(label='30 day moving average', figsize=(25,10))

# 2. ------------------>

BAC['Close'].loc['2008-01-01':'2008-12-30'].plot(label='BAC Close Price', figsize=(25,10), legend='True')

BAC['Close'].loc['2008-01-01':'2008-12-30'].rolling(30).mean().plot(label='30 day moving average', figsize=(25,10), legend='True')
```

```python
stocks_corr = bank_stocks.xs(key='Close', axis=1,level='Stock info').corr()
stocks_corr
```

```python
# Correlation between the stocks Close Price
sns.heatmap(stocks_corr,annot=True,linewidth=1, cmap=sns.cubehelix_palette(as_cmap=True))
```

```python
# Correlation between the stocks Close Price, clustermap, gruped
sns.clustermap(stocks_corr,annot=True,linewidth=1, cmap=sns.cubehelix_palette(as_cmap=True))
```

```python
# same data, using interactive cufflings
stocks_corr.iplot(kind='heatmap', colorscale='bupu')
```
